```python
print('서브 플롯 방법 -')

fig = plt.figure(figsize=(20, 7))


area01=fig.add_subplot(1,3,1)
area01.set_title('충청남도')
area01.set_xlabel('년도')
area01.set_ylabel('이동인구수')
area01.set_xticklabels(labels=seoul_frm.columns.values, rotation=90)

area01.legend(loc='best')

area02=fig.add_subplot(1,3,2)
area02.set_title('경상북도')
area02.set_xlabel('년도')
area02.set_ylabel('이동인구수')
area02.set_xticklabels(labels=seoul_frm.columns.values, rotation=90)

area02.legend(loc='best')

area03=fig.add_subplot(1,3,3)
area03.set_title('전라남도')
area03.set_xlabel('년도')
area03.set_ylabel('이동인구수')
area03.set_xticklabels(labels=seoul_frm.columns.values, rotation=90)

area03.legend(loc='best')



area01.plot(seoul_frm.columns.values, seoul_frm.loc['충청남도'], marker='+',ms=5, label='충청북도')
area02.plot(seoul_frm.columns.values, seoul_frm.loc['경상북도'], marker='+',ms=5, label='경상북도' )
area03.plot(seoul_frm.columns.values, seoul_frm.loc['전라남도'], marker='+',ms=5, label='전라남도')

plt.show()
plt.close()
```
